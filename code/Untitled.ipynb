```python
import geopandas as gpd
from shapely.geometry import Polygon

# Assuming 'water_bodies' is your GeoDataFrame with water bodies geometry

# Function to make a geometry valid by applying a small buffer
def make_valid(geometry):
    if not geometry.is_valid:
        return geometry.buffer(0)
    return geometry

# Check if water bodies geometries are valid and make them valid if not
water_bodies['geometry'] = water_bodies['geometry'].apply(make_valid)

# Now 'water_bodies' contains valid geometries
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
import shapely.plotting
from shapely.geometry import Polygon
import matplotlib.cm as cm

# Define the selected dates
selected_dates = ['2021-06-13', '2021-12-22', '2022-01-31']

# Use the values from the pivot_table_df directly for the colorbar
cax = cm.ScalarMappable(cmap='viridis')
cax.set_array(pivot_table_df.values.ravel())

# Custom Formating Function for the colorbar/Legend
def format_func_1(value, tick_number):
    return "{:.0f}0".format(value / 100)

def plot_north_arrow( ax, xT=0, yT=0, scale=1 ):
    def t_s(t,xT,yT,s):
        x,y = t
        return (xT+(x*s),yT+(y*s))
    a = [(0, 5), (0, 1), (2, 0)]
    b = [(0, 5), (0, 1), (-2, 0)]
    t_pos = (0.25,6)
    t_pos_x,t_pos_y = t_s(t_pos,xT,yT,scale)
    polygon1 = Polygon( [t_s(t,xT,yT,scale) for t in a] )
    polygon2 = Polygon( [t_s(t,xT,yT,scale) for t in b] )
    shapely.plotting.plot_polygon(polygon1, add_points=False, ax=ax, color=None, facecolor='None', edgecolor='k', linewidth=2)
    shapely.plotting.plot_polygon(polygon2, add_points=False, ax=ax, color=None, facecolor='k', edgecolor='k', linewidth=None)
    ax.text(x=t_pos_x,y=t_pos_y,s='N', fontsize='medium',
            ha='center',
            va='center',weight='bold')

# Create the facet grid of subplots``
fig, axes = plt.subplots(1, len(selected_dates), figsize=(3 * len(selected_dates), 3), dpi=100)  # Adjust dpi for resolution

# Initialize variable for colorbar
# cbar = None

# Iterate over the selected dates and plot each category on a separate subplot
for i, date in enumerate(selected_dates):
    ax = axes[i]

    # Perform spatial difference operation to remove districts with data from the lake area
    # shape_without_districts_in_lake = gpd.overlay(shape, water_bodies, how='difference')

    

    shape.plot(ax=ax, color='white', edgecolor='black')

    # Plot the water bodies with a blue color after the main shape
    water_bodies.plot(ax=ax, color='blue')  # Adjust alpha for transparency if needed
    
    # Simplify the geometries to reduce the number of vertices (adjust tolerance as needed)
    subset_shape = shape[shape['Date'] == date].copy()  # Create a copy to avoid modifying the original GeoDataFrame
    subset_shape['geometry'] = subset_shape['geometry'].simplify(tolerance=0.01)
    
    subset_shape.plot(ax=ax, cmap='plasma', legend=True, aspect='equal')
    ax.set_title(f' {date}')
    # ax.set_axis_off()
    # ax.get_yaxis().set_visible(False)
    ax.grid(visible=True, which='major', axis='both', linestyle=":")

    plot_north_arrow( ax, xT=30, yT=3, scale=0.1)
    
    # Add the scale bar
    scalebar = AnchoredSizeBar(ax.transData, 1, "1 Meter", 3)
    ax.add_artist(scalebar)

# Title
plt.suptitle('FP SERVICES FOR VOUCHER CLIENTS BY DISTRICTS', size=15, weight='bold')

# Adjust aspect ratio for equal scaling
ax.set_aspect('equal')

# Create colorbar only once for the last subplot
cbar = plt.colorbar(cax, ax=axes.ravel().tolist(), label='Service Count', format=format_func_1)
#cbar.set_ticks([])  # To prevent tick labels from overlapping
cbar.set_ticks([0, 25, 50, 75, 100])  # Set the tick positions
cbar.set_ticklabels(['0', '250', '500', '750', '1000'])  # Set the corresponding tick labels


# Adjust layout to prevent last subplot from being smaller
# plt.tight_layout(rect=[0, 0.03, 1, 0.95])
 

# Show the plot
plt.show()
```

```python
import matplotlib.dates as mdates
# Define a function to categorize services as short term or long term
def categorize_service_type(Service):
    short_term_services = ['Condom', 'Depo', 'Oral Contraceptive Pills - COC','Male Condom']
    if Service in short_term_services:
        return 'Short Term'
    else:
        return 'Long Term'

# Apply the categorize_service_type function to the 'Service' column and store the results in a new 'Type' column
combined_df_1['Type'] = combined_df_1['Service'].apply(categorize_service_type)

#df_2021_2022 =combined_df_1[combined_df_1['Date'].dt.year.isin([2021, 2022])]
df_2021_2022=df_2021_2022 = combined_df_1[combined_df_1['Date'].dt.year == 2021]

 # Group by 'Date', 'District', and 'Service,and Type', and sum the 'service_count' for long term services
Service_count_1 = df_2021_2022.groupby(['District','Date','Type'])['Service'].value_counts()
Service_count_1_df = Service_count_1.reset_index(name='service_count')
long_term_df = Service_count_1_df[Service_count_1_df['Type'] == 'Long Term']

y= long_term_df['service_count']
x= long_term_df['Date']

# Create a larger figure with more width
plt.figure(figsize=(12, 6))

plt.plot(x,y,'b.-')
plt.xlabel('Date')
plt.ylabel('No of FP Services')
plt.title('Long Term Services')

# Set the locator and formatter for the x-axis to display 7-month intervals
locator = mdates.MonthLocator(interval=7)
formatter = mdates.DateFormatter('%b %Y')
plt.gca().xaxis.set_major_locator(locator)
plt.gca().xaxis.set_major_formatter(formatter)
plt.show()

```

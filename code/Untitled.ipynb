```python
import geopandas as gpd
from shapely.geometry import Polygon

# Assuming 'water_bodies' is your GeoDataFrame with water bodies geometry

# Function to make a geometry valid by applying a small buffer
def make_valid(geometry):
    if not geometry.is_valid:
        return geometry.buffer(0)
    return geometry

# Check if water bodies geometries are valid and make them valid if not
water_bodies['geometry'] = water_bodies['geometry'].apply(make_valid)

# Now 'water_bodies' contains valid geometries
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import shapely.plotting
from shapely.geometry import Polygon
import matplotlib.cm as cm
from matplotlib_scalebar.scalebar import ScaleBar

# Define the selected dates
selected_dates = ['2021-06-13', '2021-12-22', '2022-01-31']

# Use the values from the pivot_table_df directly for the colorbar
cax = cm.ScalarMappable(cmap='viridis')
cax.set_array(pivot_table_df.values.ravel())

# Custom Formating Function for the colorbar/Legend
def format_func_1(value, tick_number):
    return "{:.0f}0".format(value / 100)

def plot_north_arrow(ax, xT=0, yT=0, scale=1):
    def t_s(t, xT, yT, s):
        x, y = t
        return (xT + (x * s), yT + (y * s))

    a = [(0, 5), (0, 1), (2, 0)]
    b = [(0, 5), (0, 1), (-2, 0)]
    t_pos = (0.25, 6)
    t_pos_x, t_pos_y = t_s(t_pos, xT, yT, scale)
    polygon1 = Polygon([t_s(t, xT, yT, scale) for t in a])
    polygon2 = Polygon([t_s(t, xT, yT, scale) for t in b])
    shapely.plotting.plot_polygon(polygon1, add_points=False, ax=ax, color=None, facecolor='None', edgecolor='k', linewidth=2)
    shapely.plotting.plot_polygon(polygon2, add_points=False, ax=ax, color=None, facecolor='k', edgecolor='k', linewidth=None)
    ax.text(x=t_pos_x, y=t_pos_y, s='N', fontsize='medium', ha='center', va='center', weight='bold')

# Create the facet grid of subplots
fig, axes = plt.subplots(1, len(selected_dates), figsize=(3 * len(selected_dates), 3), dpi=100)

# Iterate over the selected dates and plot each category on a separate subplot
for i, date in enumerate(selected_dates):
    ax = axes[i]

    shape.plot(ax=ax, color='white', edgecolor='black')

    water_bodies.plot(ax=ax, color='blue')  # Adjust alpha for transparency if needed

    subset_shape = shape[shape['Date'] == date].copy()
    subset_shape['geometry'] = subset_shape['geometry'].simplify(tolerance=0.01)

    subset_shape.plot(ax=ax, cmap='plasma', legend=True, aspect='equal')
    ax.set_title(f' {date}')
    ax.grid(visible=True, which='major', axis='both', linestyle=":")

    plot_north_arrow(ax, xT=30, yT=3, scale=0.1)

    # Add the custom scale bar
    scalebar = ScaleBar(1, location='lower left', length_fraction=0.1, scale_loc='bottom', units='km')
    ax.add_artist(scalebar)

# Title
plt.suptitle('FP SERVICES FOR VOUCHER CLIENTS BY DISTRICTS', size=15, weight='bold')

# Adjust aspect ratio for equal scaling
ax.set_aspect('equal')

# Create colorbar only once for the last subplot
plt.colorbar(cax, ax=axes.ravel().tolist(), label='Service Count', format=format_func_1)

# Show the plot
plt.show()
```

```python
import matplotlib.dates as mdates
# Define a function to categorize services as short term or long term
def categorize_service_type(Service):
    short_term_services = ['Condom','HIV Visit 1', 'Depo', 'Oral Contraceptive Pills - POP','Oral Contraceptive Pills - COC','Female Condom','Male Condom','Emergency Contraceptive Pills','Intramuscular Injectable services 3 month','MA/MPAC','MSP/PAC','Subcutaneous Injectable Services 3 month','Female Sterilisation/Marie Stopes Ligation (MSL)','Intramuscular Injectable services 3 month - Second Visit','Intramuscular Injectable services 3 month - Third Visit','Side Effect Management']
    if Service in short_term_services:
        return 'Short Term'
    else:
        return 'Long Term'

# Apply the categorize_service_type function to the 'Service' column and store the results in a new 'Type' column
filtered_df.loc[:, 'Type'] = filtered_df['Service'].apply(categorize_service_type)

# Group by month and calculate the sum of service counts for each month
Service_count_1 = filtered_df.groupby(['District','Date','Type'])['Service'].value_counts()
Service_count_1_df = Service_count_1.reset_index(name='service_count')
# Service_count_1_df
# Group by month, 'Service', and 'Type' and calculate the sum of service counts for each combination
grouped_by_month_service_type = Service_count_1_df.groupby([pd.Grouper(key='Date', freq='M'), 'Service', 'Type'])['service_count'].sum().reset_index()
grouped_by_month_service_type = grouped_by_month_service_type.reset_index(drop=True)
grouped_by_month_service_type


# long term services
long_term_df = grouped_by_month_service_type[grouped_by_month_service_type['Type'] == 'Long Term']
short_term_df = grouped_by_month_service_type[grouped_by_month_service_type['Type'] == 'Short Term']
#  long_term_df[long_term_df['Service'] == 'Marie Stopes Vasectomy (MSV)']
 

# Group by month and calculate the total service count for 'Long Term' services
# long_term_service_count = long_term_df.groupby(pd.Grouper(key='Date', freq='M'))['service_count'].sum()
# long_term_service_count.values




```

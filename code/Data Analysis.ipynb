```python
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.ticker import FuncFormatter
import pandas as pd

shape_path = r'D:\Work\Python\Data\shape\shape\uganda_districts_2019-wgs84.shp'

shape = gpd.read_file(shape_path)

shape.rename(columns={'DName2019': 'District'}, inplace=True)

# Merge the shape DataFrame with the pivot_table_df using 'DName2018' as the key
#shape = pd.merge(left=shape, right=pivot_table_df, left_on='DName2018', right_on='District', how='left')
shape = pd.merge(shape, summary_df, on='District', how='outer')

# Load the natural water bodies data into a new GeoDataFrame
water_bodies_path = r'D:\Work\Python\Data\shape\shape\lk.shp'
water_bodies = gpd.read_file(water_bodies_path)


# Create the plot
fig, ax = plt.subplots()

# Custom Formating Function for the colorbar/Legend
def format_func_1(value, tick_number):
    return "{:.0f}0".format(value / 100)

# Set the invisible axes
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

# Plot the areas with data
shape_plot = shape.plot(ax=ax, cmap='plasma', legend=True)

# Use the values from the pivot_table_df directly for the colorbar
cax = cm.ScalarMappable(cmap='viridis')
cax.set_array(pivot_table_df.values.ravel())

# Title
ax.set_title('FP SERVICES FOR VOUCHER CLIENTS BY DISTRICTS', size=18, weight='bold')

# Add the colorbar to the plot
plt.colorbar(cax, ax=ax, label='Service Count', format=format_func_1)

# Get the districts covered by pivot_table_df
districts_with_data = pivot_table_df.index.tolist()

# Filter out the districts that have data, leaving the ones with no data (blank areas)
blank_shape = shape[~shape['District'].isin(districts_with_data)]

# Plot the blank areas with white color
blank_shape.plot(ax=ax, facecolor='white', edgecolor='grey')

# Plot the water bodies with a blue color after the main shape
water_bodies.plot(ax=ax, color='blue')  # Adjust alpha for transparency if needed


plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar

# Assuming 'shape' is your merged GeoDataFrame with the 'Date' category
# ...

# Define the selected dates
selected_dates = ['2021-01-04', '2021-01-05', '2023-05-17']

# Create the facet grid of subplots
fig, axes = plt.subplots(1, len(selected_dates), figsize=(3 * len(selected_dates), 3), dpi=100)  # Adjust dpi for resolution

# Iterate over the selected dates and plot each category on a separate subplot
for i, date in enumerate(selected_dates):
    ax = axes[i]

    # Perform spatial difference operation to remove districts with data from the lake area
    shape_without_districts_in_lake = gpd.overlay(shape, water_bodies, how='difference')

    shape_without_districts_in_lake.plot(ax=ax, color='white', edgecolor='black')

    # Plot the water bodies with a blue color after the main shape
    water_bodies.plot(ax=ax, color='blue')  # Adjust alpha for transparency if needed
    
    # Simplify the geometries to reduce the number of vertices (adjust tolerance as needed)
    subset_shape = shape[shape['Date'] == date].copy()  # Create a copy to avoid modifying the original GeoDataFrame
    subset_shape['geometry'] = subset_shape['geometry'].simplify(tolerance=0.01)
    
    subset_shape.plot(ax=ax, cmap='plasma', legend=True, aspect='equal')
    ax.set_title(f' {date}')
    #ax.set_axis_off()
    ax.get_yaxis().set_visible(False)
    ax = plt.gca()
    scalebar = AnchoredSizeBar(ax.transData, 1, "1 Meter", 3)
    ax.add_artist(scalebar)


# Title
plt.suptitle('FP SERVICES FOR VOUCHER CLIENTS BY DISTRICTS', size=15, weight='bold')



# Show the plot
plt.show()
```
